```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
import seaborn as sns
import numpy as np
from IPython.display import display
```

```python
plot_df_1 = pd.read_csv('/Users/daviddornig/Documents/Master_Thesis/Bioinf/Code/philipp-trinh/KOOPOMICS/input_data/pregnancy/pregnancy_preprocessed.csv')
plot_df_2 = pd.read_csv('/Users/daviddornig/Documents/Master_Thesis/Bioinf/Code/philipp-trinh/KOOPOMICS/input_data/pregnancy/pregnancy_interpolated_preprocessed.csv')
```

```python
# Group by Subject ID
df_grouped = plot_df_1.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')

def display_subject_df(subject_id):
    df_subject = df_grouped.get_group(subject_id)
    df_days = pd.DataFrame(df_subject['Gestational age (GA)/days'])
    df_days['Gestational age (GA)/days difference'] = df_days['Gestational age (GA)/days'].diff()
    df_days['Subject ID'] = subject_id
    return df_days



# Sort the unique Subject IDs
sorted_subject_ids = sorted(plot_df_1['Subject ID'].unique())

# Create a dropdown widget
subject_id_dropdown = widgets.Dropdown(
    options=sorted_subject_ids,
    description='Subject ID:',
    disabled=False,
)

# Link the dropdown widget to the display function
widgets.interact(display_subject_df, subject_id=subject_id_dropdown)
display(subject_id_dropdown)
```

```python

# Group by Subject ID
df_grouped = plot_df_1.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')
interpolated_df_grouped = plot_df_2.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')

def display_subject_df(subject_id, start_day, end_day):
    df_subject = df_grouped.get_group(subject_id)
    df_interpolated_subject = interpolated_df_grouped.get_group(subject_id)

    # Filter by the selected range of days
    df_subject_filtered = df_subject[(df_subject['Gestational age (GA)/days'] >= start_day) & (df_subject['Gestational age (GA)/days'] <= end_day)]
    df_interpolated_filtered = df_interpolated_subject[(df_interpolated_subject['Gestational age (GA)/days'] >= start_day) & (df_interpolated_subject['Gestational age (GA)/days'] <= end_day)]

    print('Original Dataframe:')
    display(df_subject_filtered)
    print('Interpolated Dataframe:')
    display(df_interpolated_filtered)

# Sort the unique Subject IDs
sorted_subject_ids = sorted(plot_df_1['Subject ID'].unique())

# Create interactive widgets
subject_id_dropdown = widgets.Dropdown(
    options=sorted_subject_ids,
    description='Subject ID:',
    disabled=False,
)

start_day_slider = widgets.IntSlider(
    value=(plot_df_2['Gestational age (GA)/days'].min()-7),
    min=(plot_df_2['Gestational age (GA)/days'].min()-7),
    max=plot_df_2['Gestational age (GA)/days'].max()+7,
    step=7,
    description='Start Day:',
    continuous_update=False
)

end_day_slider = widgets.IntSlider(
    value=(plot_df_2['Gestational age (GA)/days'].min()+7),
    min=(plot_df_2['Gestational age (GA)/days'].min()-7),
    max=plot_df_2['Gestational age (GA)/days'].max()+7,
    step=7,
    description='End Day:',
    continuous_update=False
)

# Function to update end_day_slider's min value based on start_day_slider's value
def update_end_slider(*args):
    end_day_slider.min = start_day_slider.value
    if end_day_slider.value < end_day_slider.min:
        end_day_slider.value = end_day_slider.min

start_day_slider.observe(update_end_slider, 'value')

# Link the dropdown and sliders to the display function
widgets.interact(display_subject_df, subject_id=subject_id_dropdown, start_day=start_day_slider, end_day=end_day_slider)

# Display the widgets
display(subject_id_dropdown, start_day_slider, end_day_slider)
```

```python

df_grouped = plot_df_1.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')

# Concatenate all df_days DataFrames
all_df_days = pd.concat([display_subject_df(subject_id) for subject_id in sorted_subject_ids])

# Plot combined boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Subject ID', y='Gestational age (GA)/days', data=all_df_days, palette='Set3', hue='Subject ID')
plt.title('Boxplot of Gestational age (GA)/days for Original Data')
plt.xlabel('Subject ID')
plt.ylabel('Gestational age (GA)/days')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df_grouped = plot_df_2.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')


# Concatenate all df_days DataFrames
all_df_days = pd.concat([display_subject_df(subject_id) for subject_id in sorted_subject_ids])

# Plot combined boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Subject ID', y='Gestational age (GA)/days', data=all_df_days, palette='Set3', hue='Subject ID')
plt.title('Boxplot of Gestational age (GA)/days for Interpolated Data')
plt.xlabel('Subject ID')
plt.ylabel('Gestational age (GA)/days')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Assuming df and interpolated_df are already defined dataframes

# Function to plot time series data for a given Subject ID and feature
def plot_feature_timeseries(df, interpolated_df, subject_id, feature_name, time_unit):
    # Filter data for the specified subject ID
    subject_data_original = plot_df_1[plot_df_1['Subject ID'] == subject_id]
    subject_data_interpolated = interpolated_df[interpolated_df['Subject ID'] == subject_id]

    if time_unit == 'Gestational age (GA)/weeks':
        subject_data_original = subject_data_original.sort_values(by='Gestational age (GA)/weeks')
        subject_data_interpolated = subject_data_interpolated.sort_values(by='Gestational age (GA)/weeks')
        x_label = 'Gestational age (GA)/weeks'
    else:
        subject_data_original = subject_data_original.sort_values(by='Gestational age (GA)/days')
        subject_data_interpolated = subject_data_interpolated.sort_values(by='Gestational age (GA)/days')
        x_label = 'Gestational age (GA)/days'

    # Plotting
    plt.figure(figsize=(15, 6))
    plt.plot(subject_data_original[x_label], subject_data_original[feature_name], marker='o', linestyle='-', color='g')
    plt.plot(subject_data_interpolated[x_label], subject_data_interpolated[feature_name], marker='o', linestyle='--', color='b')

    plt.title(f'Time Series Plot of {feature_name} for Subject ID {subject_id}')
    plt.xlabel(x_label)
    plt.ylabel(feature_name)
    if time_unit == 'Gestational age (GA)/days':
        plt.gca().xaxis.set_major_locator(plt.MultipleLocator(7))  # Show ticks every 7 days
    if time_unit == 'Gestational age (GA)/weeks':
        plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))  # Show ticks every week

    plt.grid(True)
    plt.legend(['Original Data', 'Interpolated Data'])
    plt.tight_layout()
    plt.show()

# Get unique Subject IDs and feature columns
sorted_subject_ids = sorted(plot_df_1['Subject ID'].unique())
feature_columns = plot_df_1.select_dtypes(include=[np.float64, np.float32]).columns

# Create dropdown widgets for Subject ID, Feature, and Time Unit selection
subject_id_dropdown = widgets.Dropdown(
    options=sorted_subject_ids,
    description='Subject ID:',
    disabled=False,
)

feature_dropdown = widgets.Dropdown(
    options=feature_columns,
    description='Feature:',
    disabled=False,
)

time_unit_dropdown = widgets.Dropdown(
    options=['Gestational age (GA)/weeks', 'Gestational age (GA)/days'],
    description='Time Unit:',
    disabled=False,
)


# Define an interactive function to plot based on dropdown selections
def plot_interactive_timeseries_plot(subject_id, feature_name, time_unit):
    plot_feature_timeseries(plot_df_1, plot_df_2, subject_id, feature_name, time_unit)

# Display the interactive widget
widgets.interact(plot_interactive_timeseries_plot,
                 subject_id=subject_id_dropdown,
                 feature_name=feature_dropdown,
                 time_unit=time_unit_dropdown)
```
