# Convert numerical columns to float

```python

for col in df.columns:
    if df[col].dtype == 'object':  # Check if column is of object (string) type
        try:
            df[col] = df[col].str.replace(',', '.').astype(float)
        except:
            pass  # Skip columns that cannot be converted to float
```

# Normalize and impute for equal time intervals

```python
# Calculate 'Gestational age (GA)/days' for easier Imputation Inspection:
df['Gestational age (GA)/days'] = df['Gestational age (GA)/weeks'] * 7

# Insert 'Gestational age (GA)/days' at position 1
df.insert(1, 'Gestational age (GA)/days', df.pop('Gestational age (GA)/days'))

metabolite_names = df.columns[6:]

agg_dict = {col: 'first' for col in df.columns if col not in metabolite_names}
agg_dict.update({col: 'mean' for col in metabolite_names})
df = df.groupby(['Subject ID', 'Gestational age (GA)/days'], as_index=False).agg(agg_dict)

df
```

```python
df_grouped = df.sort_values(['Subject ID', 'Gestational age (GA)/days']).groupby('Subject ID')
df_subject = df_grouped.get_group('DP01')  # Replace 'your_subject_id' with the actual ID
df_subject
```

```python
subject_info = []
for subject_id, _ in df_grouped:
    df_subject = df_grouped.get_group(subject_id)
    min_days = (df_subject['Gestational age (GA)/weeks'] * 7).diff().min()
    max_days = (df_subject['Gestational age (GA)/weeks'] * 7).diff().max()
    start_date = (df_subject['Gestational age (GA)/weeks'] * 7).min()
    end_date = (df_subject['Gestational age (GA)/weeks'] * 7).max()
    subject_info.append({'Subject ID': subject_id,
                        'Length': len(df_subject['Gestational age (GA)/weeks']),
                        'Min Days': min_days,
                        'Max Days': max_days,
                         'Start Date': start_date,
                         'End Date': end_date})

df_info = pd.DataFrame(subject_info)
df_info
```

```python
# Calculate the common start and end dates
common_start_date = df_info['Start Date'].min()
common_end_date = df_info['End Date'].max()

# Create a common timeline in days
common_timeline = np.arange(0, common_end_date+7, 7).astype(int)
```

```python
common_timeline
```

```python
# Find the index of the closest value greater than or equal to the start date
start_index = np.searchsorted(common_timeline, common_start_date, side='right')
if start_index > 0 and (abs(common_timeline[start_index] - common_start_date)) > (abs(common_timeline[start_index - 1] - common_start_date)):
    start_index -= 1

# Find the index of the closest value less than or equal to the end date
end_index = np.searchsorted(common_timeline, common_end_date, side='left')
if end_index < len(common_timeline) and (abs(common_timeline[start_index] - common_end_date)) > (abs(common_timeline[start_index - 1] - common_end_date)):
    end_index -= 1

# Filter the timeline
filtered_timeline = common_timeline[start_index:end_index]

print("Filtered Timeline:", filtered_timeline)
```

```python


# Function to interpolate data for a subject and count artificial data points
def interpolate_subject_data(subject_df, common_timeline):
    subject_timeline = (subject_df['Gestational age (GA)/days']).values.reshape(-1, 1)
    interpolated_data = {}
    original_data_points = np.array(sorted(subject_timeline)).flatten()
    artificial_data_points = 0

    float_columns = subject_df.select_dtypes(include=[np.float64, np.float32]).columns
    string_columns = subject_df.select_dtypes(include=[object]).columns

    interpolated_timeline = []
    for feature in float_columns:
        interpolated_values = []
        for day in common_timeline:

            if day not in original_data_points:

                # Find the nearest lower and higher data points
                lower_points = original_data_points[original_data_points < day]
                higher_points = original_data_points[original_data_points > day]

                if lower_points.size == 0 or higher_points.size == 0:
                    continue  # Skip if there are no points to interpolate from

                lower_point = lower_points.max()
                higher_point = higher_points.min()

                # Calculate the weights
                weight_lower = round(higher_point - day)
                weight_higher = round(day - lower_point)

                if weight_lower > 10 or weight_higher > 10:
                    continue  # Skip if the weights are too large

                total_weight = weight_lower + weight_higher

                # Get the values at the lower and higher points
                value_lower = subject_df.loc[subject_df['Gestational age (GA)/days'] == lower_point, feature].values[0]
                value_higher = subject_df.loc[subject_df['Gestational age (GA)/days'] == higher_point, feature].values[0]

                # Calculate the weighted average
                interpolated_value = (value_lower * weight_lower + value_higher * weight_higher) / total_weight
                interpolated_values.append(interpolated_value)
                interpolated_timeline.append(day)
            else:
                # Find the most recent data point before the current day
                previous_points = original_data_points[original_data_points < day]
                previous_point = previous_points.max() if previous_points.size > 0 else np.nan

                # Find the closest data point after the current day
                next_points = original_data_points[original_data_points > day]
                next_point = next_points.min() if next_points.size > 0 else np.nan

                # Check if the difference with the previous data point is within 7 days
                if not pd.isna(previous_point) and (day - previous_point) <= 7:
                    feature_value = subject_df.loc[subject_df['Gestational age (GA)/days'] == day, feature].values
                    if feature_value.size > 0:
                        interpolated_values.append(feature_value[0])

                # Check if the difference with the next data point is within 7 days
                if not pd.isna(next_point) and (next_point - day) <= 7:
                    feature_value = subject_df.loc[subject_df['Gestational age (GA)/days'] == day, feature].values
                    if feature_value.size > 0:
                        interpolated_values.append(feature_value[0])

        for col in string_columns:
          interpolated_data[col] = subject_df[col].iloc[0]

        interpolated_data[feature] = interpolated_values

    # Calculate the number of artificial data points
    for day in common_timeline:
        if day not in original_data_points:
            artificial_data_points += 1

    interpolated_df = pd.DataFrame(interpolated_data)

    total_points = len(common_timeline)
    percent_artificial = (artificial_data_points / total_points) * 100

    return interpolated_df, artificial_data_points, total_points, percent_artificial

# Apply interpolation and calculate artificial data for each subject
interpolated_subjects = []
artificial_data_summary = []

for subject_id in df_info['Subject ID'].unique():
    subject_df = df[df['Subject ID'] == subject_id]
    interpolated_df, artificial_data_points, total_points, percent_artificial = interpolate_subject_data(subject_df, common_timeline)
    interpolated_subjects.append(interpolated_df)
    artificial_data_summary.append({
        'Subject ID': subject_id,
        'Artificial Data Points': artificial_data_points,
        'Total Points': total_points,
        'Percent Artificial Data': percent_artificial
    })

# Concatenate all interpolated data
interpolated_df = pd.concat(interpolated_subjects, ignore_index=True)
artificial_data_summary_df = pd.DataFrame(artificial_data_summary)
```
